```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('istherecorrelation.csv', delimiter = ';', decimal=",")

fig, ax1 = plt.subplots()
ax1.plot(data["Year"],data['WO [x1000]'], label = 'WO')
ax2 = ax1.twinx()
ax2.plot(data["Year"],data['NL Beer consumption [x1000 hectoliter]'], label = 'Beer consumption', c = 'orange')
ax1.legend(loc=0), ax2.legend(loc='upper right')
ax1.set(xlabel = "Year", ylabel = "WO [x1000]"), ax2.set(ylabel = "NL Beer consumption [x1000 hectoliter]") 
plt.rcParams['figure.dpi'] = 300
plt.show()
plt.savefig('visualization.jpg')
```
